```
#TRANSFORMANDO STR PYTON EM JSON OBJECT
people_string = '''
{
  "people": [
    {
      "name": "Daniel",
      "phone": "98225-7991",
      "emails": ["dmca@cin.ufpe.br", "danielmoraescostaandrade@gmail.com"],
      "has_license": false
    },
    {
      "name": "Fernanda",
      "phone": "98533-8710",
      "emails": null,
      "has_license": true
    }
  ]
}
''' #string em python



data = json.loads(people_string) #varaivel adquirida, transformando em dicionario
print(data)
print(type(data))
print(type(data["people"]))
```

```
#FOR PARA PRINT
for person in data['people']: #printando as pessoas da lista que contem dicionarios
  print(person)
  print(person['name']) #cada nome
```

```
#FOR PARA PRINT DELETANDO ALGUMA CHAVE
for person in data['people']: #printando as pessoas da lista que contem dicionarios
  del person['phone']
  print(person)

print()
#TRANSFORMANDO JSON OBJECT EM PYTHON DNV
new_string = json.dumps(data, indent=2) #,sort_keys= True se quiser ordem das chaves alfabeticas
print(new_string)
print(type(new_string)) #classe str
```

```
#TRABALHANDO COM ARQUIVO JSON EXTERNP
with open('states.json') as f: #abriu o arquivo json
  data = json.load(f) #variavel atribuida 

  for state in data['states']: #printando todos os dicionarios de estados
    print(state)
```

```
#PRINTANDO DADOS DO ARQUIVO JSON
with open('states.json') as f: #abriu o arquivo json
  data = json.load(f) #variavel atribuida 

  for state in data['states']: #printando todos os dicionarios de estados
    print(state['name'], state['abbreviation']) #printa só os nomes dos estados e sigla
```

```
#PRINTANDO DADOS DO ARQUIVO JSON (EXCETO A ABREVIACAO)
with open('states.json') as f: #abriu o arquivo json
  data = json.load(f) #variavel atribuida 

  for state in data['states']: #printando todos os dicionarios de estados
    del state['abbreviation']
    print(state) #printa só os nomes dos estados e area codes
```

```
#CRIANDO UM NOVO ARQUIVO JSON SEM O AREA CODE
with open('states.json') as f: #abriu o arquivo json
  data = json.load(f) #variavel atribuida 
  for state in data['states']: #printando todos os dicionarios de estados
    del state['area_codes']

  with open('new_states.json', 'w') as f:
    json.dump(data, f, indent=2)
  
```

```
#TRABALHANDO COM URL COM JASON
import json
from urllib.request import urlopen

with urlopen("http://maps.googleapis.com/maps/api/geocode/json?address=google") as response:
    source = response.read()
#if print(source) vai aparecer uma string enorme do conteudo da pagina

data = json.loads(source) #transformando para um objeto em pytjon atribuindo a variavel

#print(json.dumps(data, indent=2)) vai aparecer o print do data porem organizado pelo indent


usd_rates = dict() #criei o dicionario para variavel

for item in data['list']['resources']: #filtrando o dicionario para trabalhar só com os resources
    name = item['resource']['fields']['name'] #printando nome
    price = item['resource']['fields']['price'] #printando preço
    usd_rates[name] = price

print(50 * float(usd_rates['USD/INR'])) #conversao de 50 dolares para certa moeda
```

```
import json
import requests

base_url ="http://api.postmon.com.br/v1/cep/%s"

def cepFinder(cep): ## retorna em dicionario
  response = requests.get(base_url % cep)
  data = response.json()
  return data

cep = input('Digite o cep:\n')
x = cepFinder(cep)

print('\n\nRua : '+ x['logradouro'])
print('Bairro : '+ x['bairro'])
print('Cidade : '+ x['cidade'])
print('Estado : '+ x['estado'])


print('\n\n\n json file:\n\n')
print(x)
```
