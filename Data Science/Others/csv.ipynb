```
import csv
```

```
"""PRINTANDO TODAS AS LINHAS CSV"""
with open('names.csv', 'r') as csv_file: #abrir o arquivo csv
    ler_csv = csv.reader(csv_file) #variavel para ler o arquivo
    next(ler_csv) #pular a primeira linha
    for line in ler_csv:#for para printar todas as linhas do csv (a primeira linha é o field name)
      print(line) #para printar linha por linha
    
```

```
"""PRINTANDO DETERMINADA COLUNA CSV"""
with open('names.csv', 'r') as csv_file: #abrir o arquivo csv
    ler_csv = csv.reader(csv_file) #variavel para ler o arquivo

    for line in ler_csv:#for para printar todas as linhas do csv (a primeira linha é o field name)
      print(line[2]) #para printar apenas o email do csv
```

```
"""LENDO E ESCREVENDO ARQUIVO CSV"""
with open('names.csv', 'r') as csv_file: #abrir o arquivo csv
    ler_csv = csv.reader(csv_file) #variavel para ler o arquivo

    with open('new_names.csv', 'w') as new_file:   #criando um novo csv
        escrever_csv = csv.writer(new_file, delimiter='\t')    #variavel para escrever o arquivo, determinando o delimitador

        for line in ler_csv: #para escrever um novo csv
            escrever_csv.writerow(line) #escrever linha por linha
```

```
"""PRINTANDO TODAS AS LINHAS"""
with open('new_names.csv', 'r') as csv_file: #abrir o arquivo csv
    ler_csv = csv.reader(csv_file, delimiter='\t') #variavel para ler o arquivo e avisando que o delimitador é \t para n printar o \t
    for line in ler_csv: #for para printar line
        print(line) 
 
```

```
"""USANDO DICT PARA LER CSV"""
with open('names.csv', 'r') as csv_file: #abrindo o arquivo
    ler_csv = csv.DictReader(csv_file) #variavel para ler o arquivo 
    
    for line in ler_csv: #for para ler linha por linha
        print(line) #printa linha
```

```
"""USANDO DICT PARA LER CSV EM DETERMINADA COLUNA"""
with open('names.csv', 'r') as csv_file: #abrindo o arquivo
    ler_csv = csv.DictReader(csv_file) #variavel para ler o arquivo 
    
    for line in ler_csv: #for para ler linha por linha
        print(line['email']) #printa email apenas
```

```
"""USANDO DICT PARA ESCREVER UM ARQUIVO"""
with open('names.csv', 'r') as csv_file: #abrindo o arquivo
    ler_csv = csv.DictReader(csv_file) #variavel para ler o arquivo

    with open('new_names.csv', 'w') as new_file:  # criando um novo csv

        field_names = ['first_name', 'last_name', 'email'] #definindo os field names

        escrever_csv = csv.DictWriter(new_file, fieldnames=field_names, delimiter='\t')  # variavel para escrever o arquivo, determinando o delimitador e os fn

        escrever_csv.writeheader() #escreve os fieldnames

        for line in ler_csv:  # para escrever um novo csv
            escrever_csv.writerow(line)  # escrever linha por linha
```

```
  """ELIMINAR UM FIELDNAME """
  
  with open('names.csv', 'r') as csv_file: #abrindo o arquivo
    ler_csv = csv.DictReader(csv_file) #variavel para ler o arquivo
    with open('new_names.csv', 'w') as new_file:  # criando um novo csv
          field_names = ['first_name', 'last_name'] #definindo os field names sem o 'email'

          escrever_csv = csv.DictWriter(new_file, fieldnames=field_names, delimiter='\t')  # variavel para escrever o arquivo, determinando o delimitador e os fn

          escrever_csv.writeheader() #escreve os fieldnames

          for line in ler_csv:  # para escrever um novo csv
              del line['email'] #delatando a line email
              escrever_csv.writerow(line)  # escrever linha por linha
```

```
with open('persons.txt') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            print(f'\t{row[0]} works in the {row[1]} department, and was born in {row[2]}.')
            line_count += 1
    print(f'Processed {line_count} lines.')
```
