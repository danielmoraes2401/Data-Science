```
import requests
import pandas as p
```

```
#Agora, nós temos um objeto Response chamado response. Nós podemos pegar todas as informações que precisamos desse objeto.
response = requests.get('https://xkcd.com/353/')

print(dir(response)) #diretorio
```

```
#Agora, nós temos um objeto Response chamado response. Nós podemos pegar todas as informações que precisamos desse objeto.
response = requests.get('https://xkcd.com/353/')
print(response.text) #printa o source da pag
```

```
#download image usando python
response = requests.get('https://imgs.xkcd.com/comics/python.png')
with open('comic.png', 'wb') as f:
    f.write(response.content) #salvar essa png no msm diretorio
```

```
#codigo de resposta http
'''
Respostas de informação (100-199),
Respostas de sucesso (200-299),
Redirecionamentos (300-399)
Erros do cliente (400-499)
Erros do servidor (500-599).
'''
response = requests.get('https://xkcd.com/353/')
print(response.status_code)
print(response.ok) #verifica se ta -400
```

```
#Nós podemos visualizar os cabeçalhos da resposta do servidor usando um dicionário
response = requests.get('https://imgs.xkcd.com/comics/python.png')
print(response.headers) 
```
