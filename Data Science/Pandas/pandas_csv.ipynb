```
"""ACESSAR UM ESPECIFICO DADO POR CAUSA DO BOOL"""
#Access a group of rows and columns by label(s) or a boolean array.
df = pd.read_csv('pokemon_data.csv')

filtro = df.loc[ (df['Type 1'] == 'Grass') & (df['Speed'] >= 97)] #acessar onde no data frame o type 1 é igual à Fire e a velocidade é maior ou igual a 97

filtro = filtro.reset_index(drop=True) # resetando o index
filtro
```

```
"""FILTERING"""
df = pd.read_csv('pokemon_data.csv')
# se quiser printar os restos (~)
filt = (df['Speed'] >= 70) & (df['Type 1'] == 'Grass') #Verificando em forma de bool quais pokemons tem a speed > 50 e adquirindo uma variavel

#df[filt] #cria  um novo dataFrame apenas com a row que recebeu True ou direto Ex: df [ df['Speed'] >= 70] or Ex: df.loc[filt]

df.loc[filt, ['Speed', 'Type 1']] #filtrar apenas a coluna Speed e Type 1
```

```
"""FILTERING"""
df = pd.read_csv('pokemon_data.csv')
filt = df[['Name', 'Speed']][df['Speed'] == df['Speed'].min()] #filtrando só a row de maior Speed, mostrando só o Nome é a velocidade
filt
```

```
"""FILTERING WITH A LIST"""
df = pd.read_csv('pokemon_data.csv')
pokemons = ['Bulbasaur', 'Squirtle', 'Fearow', 'Pikachu', 'Onix'] #lista de pokemons
filt = df['Name'].isin(pokemons) #variavel filtrando apenas os pokemons da lista

df.loc[filt, 'Speed']
```
